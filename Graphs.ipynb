```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the graph
data = np.array([[10, 20, 30, 10, 20, 30], [15, 25, 35, 10, 20, 30], [12, 18, 28, 10, 20, 30]]), 

# Define the color scheme
colors = ['r', 'g', 'b', 'c', 'm', 'y']

# Define the categories
categories = ['Category 1', 'Category 2', 'Category 3']

# Set the width of each bar
bar_width = 0.2

# Set the position of each bar on the x-axis
bar_positions = np.arange(len(categories))

# Create the bar graph
for i in range(len(colors)):
    plt.bar(bar_positions + i*bar_width, data[:,i], width=bar_width, color=colors[i], edgecolor='black')

# Add labels and titles
plt.xticks(bar_positions + bar_width, categories)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph with 6 Colors and 3 Categories')

# Show the graph
plt.show()
```
